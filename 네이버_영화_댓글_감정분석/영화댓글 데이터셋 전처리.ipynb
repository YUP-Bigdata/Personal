```python
from autograd import numpy
from autograd import grad
from matplotlib import pyplot
import pandas as pd
```

```python
df = pd.read_csv("samples.csv")
```

```python
df.info()
```

```python
df.score
```

```python
bins = (0, 5, 10)
group_names = ['bad', 'good']
df['score'] = pd.cut(df['score'], bins = bins, labels = group_names)
```

```python
df.score
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

#Now lets assign a labels to our quality variable
label_quality = LabelEncoder()
```

```python
df['score'] = label_quality.fit_transform(df['score'])
```

```python
df.score
```

```python
x = df
y = df.score

# 훈련데이터 세트 분할
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.2, random_state=42)
```

```python
x_train.drop(['score'], axis=1, inplace=True)
x_test.drop(['score'], axis=1, inplace=True)
```

```python
y_train
```

```python
train = pd.DataFrame({
        "comment": x_train["sentence"],
        "good/bad": y_train
    })
```

```python
test = pd.DataFrame({
        "comment": x_test["sentence"],
        "good/bad": y_test
    })
```

```python
train.to_csv('train.csv',index = False)
```

```python
test.to_csv('test.csv',index = False)
```

```python
y_test.to_csv('y_test.csv',index = False)
```
